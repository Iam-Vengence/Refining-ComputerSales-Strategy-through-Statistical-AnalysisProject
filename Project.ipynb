**a) Analyze Purchase Likelihood Based on Specifications**



```python


import numpy as np
import pandas as pd
data = pd.read_csv(r'/content/DS1_C8_Computers_Data_Project.csv')
filtered_data = data[(data['ram'] >= 8) & (data['screen'] == 14) & (data['speed'] > 33)]
total_computers = len(data)
computers_meeting_criteria = len(filtered_data)
probability = computers_meeting_criteria / total_computers
print(f"Total number of computers: {total_computers}")
print(f"Number of computers meeting the criteria: {computers_meeting_criteria}")
print(f"Probability of purchasing such a computer: {probability:.2f}")

# The data is cleaned before perfroming tasks the probability of purchasing such a computer is 0.16
```

**b) Probability of Selling Premium Computers**

```python

# |)
premium_computers = data[(data['premium'] == 'yes')]
number_of_premiumcomputers = len(premium_computers)
probability_1 = number_of_premiumcomputers/total_computers
print(f"Total number of premium computers: {number_of_premiumcomputers}")
print(f"Number of computers meeting the criteria: {number_of_premiumcomputers}")
print(f"Probability of purchasing such a computer: {probability_1:.2f}")

# The probability of purchasing such a computer is 0.90 is calculated by filtering the data by yes in premiun
```

```python

# ||)
premium_cd_computers = data[(data['premium'] == 'yes')&(data ['cd']=='yes')]
no_of_premium_cd_computers = len(premium_cd_computers)
probability_of_premiumcdcomputers = no_of_premium_cd_computers/number_of_premiumcomputers


Probility_of_selecting_premium_cd_computers = (probability_1*probability_of_premiumcdcomputers)/probability_1

print(f"Probability of purchasing such a premium computer with cd player: {Probility_of_selecting_premium_cd_computers:.2f}")

# The task is done using conditional probaility becuse the cd should be seleccted after the premium computer is selected. so, the intersection of cd and premium divided by probality of premium computer.
# The rchasing such a computer is 0.50
```

```python
# a)
Total_price = data['price'].mean()
print(f"Average totalprice:{Total_price:.2f}")

# The avergare total price of computers or population mean of computer price is 2219.58
```

```python
# b)
# random sampling
import numpy as np
dat1 = data['price']
a = dat1.sample(10,replace=False).mean()
print('randam sample price mean:',a)


# Stratified Sampling
premgrp = data.groupby('premium')
premium = data['premium'].unique()
str_sam = pd.DataFrame()
for i in premium:
  d_data = premgrp.get_group(i)
  sample_data = d_data.sample(5)
  str_sam = pd.concat([str_sam, sample_data])

str_sam.reset_index(drop=True,inplace=True)
b=str_sam['price'].sum()
print('Stratified sample mean:',b)




# cluster sampling
sub_grp = data.groupby('cd')
sub = ['yes','no']
clu_sam = pd.DataFrame()
for i in sub:
  d_data = sub_grp.get_group(i)
  sample_data1 = d_data.sample(5)
  clu_sam = pd.concat([clu_sam, sample_data1])

clu_sam.reset_index(drop=True,inplace=True)
c=clu_sam['price'].sum()
print('Cluster sample mean:',c)


# systematic smapling
dq= data[::600]
d= dq['price'].mean()
print(f"Systematic sample mean:{d:.2f}")


# The mean in each type of sample is different from other ramdomsample mean and cluster sample have little difference , stratified sample and systemic have little difference
# but eventually there are differences in each different sample mean
```

```python
# c)
premiumcom_totalprice = data.loc[data['premium']=='yes']
non_premiumcom_totalprice= data.loc[data['premium']=='no']
average_priceof_precomputers = premiumcom_totalprice['price'].mean()
average_priceof_non_precomputers = non_premiumcom_totalprice['price'].mean()
s100 = premiumcom_totalprice.sample(100,replace=False,random_state=1).sort_index()
n100 = non_premiumcom_totalprice.sample(100,replace=False,random_state=1).sort_index()

print(f"Aveage price of premium computers: {average_priceof_precomputers:.2f} ")
print(f"Aveage price of 100 premium computers samples: {s100['price'].mean():.2f} ")
print(f'Average price of non premium computers:{average_priceof_non_precomputers:.2f}')
print(f"Average price of 100 non premium computers samples:{n100['price'].mean():.2f}")

# The means of prices of premium and its sample ,non premium and its sample have so light difference it can be said its lowly approx same may be it has a slight difference in plots
# It can be said the sample i have got match almost the  same charatertstics as the orginal population data.
```

```python
# a)
import matplotlib.pyplot as plt
import seaborn as sb
fig,ax = plt.subplots(1,5,figsize=(14,7))
data1=data['price']
s10 = data1.sample(10,replace=False,random_state=1).sort_index()
sb.histplot(data= s10,ax=ax[0],kde=True)
s50 = data1.sample(50,replace=False,random_state=1).sort_index()
sb.histplot(data= s50,ax=ax[1],kde = True)
s200 = data1.sample(200,replace=False,random_state=1).sort_index()
sb.histplot(data= s200,ax=ax[2],kde = True)
s1000 = data1.sample(1000,replace=False,random_state=1).sort_index()
sb.histplot(data= s1000,ax=ax[3],kde = True)
s3000 = data1.sample(3000,replace=False,random_state=1).sort_index()
sb.histplot(data= s3000,ax=ax[4],kde = True)

# The distribution of a sample change by increase in sample size and from the above damples distribution its seen that the distibution nearly becomes normal distribution as size increases
```

```python
# b)
fig,ax=plt.subplots(1,2,figsize=(14,7))
s1000 = data1.sample(1000,replace=False,random_state=1).sort_index()
sb.histplot(data= s1000,ax=ax[0],kde = True)
s3000 = data1.sample(3000,replace=False,random_state=1).sort_index()
sb.histplot(data= s3000,ax=ax[1],kde = True)


# Yes,the distribution of sample means approximate a normal distribution as sample size increases
# The visualization clearly visualize the distribution of high sample sizes is nearly normal
```

```python
# c)
sample_size2= [3000]
num_samples=3000
for i,j in enumerate(sample_size2):
  sample_means1=[]
  for _ in range(num_samples):
    sample = data1.sample(n=j,replace=True)
    sample_means1.append(np.mean(sample))
  plt.hist(sample_means1,bins=100,color='orange',edgecolor = 'black',density=True)
  plt.title('Samplespace_3000')
  plt.xlabel('Sample_Mean')
  plt.ylabel('Frequency')

plt.show()

# Yes properties of Central limit theorem for the samples is satisfied it can be confrim from below distributed visualization of sample of size 3000.
```
